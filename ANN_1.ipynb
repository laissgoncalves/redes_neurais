```python
>>> from sklearn.neural_network import MLPClassifier
>>> from sklearn.datasets import make_classification
>>> from sklearn.model_selection import train_test_split
>>> X, y = make_classification(n_samples=100, random_state=1)
>>> X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,
...                                                     random_state=1)
```

 # HiperParâmetros :

  a) Learning rate - taxa de aprendizado , fator que multiplica o gradiente (atualização dos pesos)

  b) Hidden Size - número de neurônios na camada intermediária ; número de camadas

  c) Activation - função de mapeamento de cada neurônio

  d) Epochs - época representa uma passada por todo o conjunto de treino
   
    d.1) Batch size - tamanho dos subconjuntos de treinos que serão apresentados à rede neural (Batelada)

```python
clf = MLPClassifier(random_state=1,hidden_layer_sizes= (20,),learning_rate_init=0.001, max_iter=500,
                    verbose=True).fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
#plt.plot(clf.validation_scores_)
plt.plot(clf.loss_curve_)
#plt.plot(clf.validation_scores_)
plt.show()
```

```python
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA().fit(X_train)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python
pca = PCA(n_components=0.85)
pca.fit(X_train)
X_pca_Train = pca.transform(X_train)
```

```python
clf =  MLPClassifier(random_state=1, max_iter=300,
                    verbose=True).fit(X_pca_Train, y_train)
```

```python
X_pca_Test = pca.transform(X_test)
clf.score(X_pca_Test,y_test)
```

```python
## Normalization ###
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```
