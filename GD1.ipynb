```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.arange(0,3,0.01)
```

```python
## função custo ####
y = x**4 -3*x**3 +2
```

```python
plt.plot(x,y)
plt.grid()
plt.axhline(y=0, color="black", linestyle="--")
plt.axvline(x=2.25, color="black", linestyle="--")
```

```python
y_prime = 4*x**3 - 9*x**2
```

```python
plt.plot(x,y_prime)
plt.grid()
plt.axhline(y=0, color="black", linestyle="--")
plt.axvline(x=2.25, color="black", linestyle="--")
```

```python
x_old=0
x_new=6
eps = 0.01
precision = 0.000001
def f_prime(x):
  return 4*x**3 - 9*x**2

while abs(x_new-x_old) > precision:

  x_old=x_new
  x_new=x_old - eps*f_prime(x_old)
  print(x_new)

```
